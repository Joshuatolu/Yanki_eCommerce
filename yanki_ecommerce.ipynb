##### Data Cleaning and Transformation

```python
# Drop missing values - order_id, customer_id, and customer_name
yanki_data.dropna(subset=['Order_ID', 'Customer_ID'], inplace=True) # drop all the records that have missing order_id and customer_id
```

```python
yanki_data.info()
```

```python
# Convert Order_Date to datetime data type
yanki_data['Order_Date'] = pd.to_datetime(yanki_data['Order_Date'])
```

```python
yanki_data.info()
```

```python
# Customer Table
# First let's create the columns for our first name and last name

yanki_data[['First_Name', 'Last_Name']] = yanki_data['Customer_Name'].str.split(' ', n=1, expand=True) # n=1 is to split on the first occurence of a space
```

```python
yanki_data.columns
```

```python
# Now lets create the customer table
customer_tb = yanki_data[['Customer_ID', 'First_Name', 'Last_Name', 'Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
customer_tb.head()
# customer_tb.info()
```

```python
# Products Table
product_tb = yanki_data[['Product_ID', 'Product_Name', 'Brand', 'Category', 'Price']].copy().drop_duplicates().reset_index(drop=True)
product_tb.info()
```

```python
# Shipping Table
shipping_tb = yanki_data[['Postal_Code', 'Shipping_Address', 'City', 'State', 'Country', 'Customer_ID']].copy().drop_duplicates().reset_index(drop=True)
shipping_tb['Shipping_ID'] = range(1, len(shipping_tb)+1)
shipping_tb = shipping_tb[['Shipping_ID', 'Customer_ID', 'Postal_Code', 'City', 'State', 'Country', 'Shipping_Address']]
shipping_tb.head()
```

```python
# Order Table
order_tb = pd.merge(yanki_data[['Order_ID', 'Customer_ID', 'Product_ID', 'Quantity', 'Total_Price', \
                    'Order_Date', 'Payment_Method']], shipping_tb[['Customer_ID', 'Shipping_ID']],\
                    on='Customer_ID').copy().drop_duplicates().reset_index(drop=True)
order_tb = order_tb[['Order_ID', 'Customer_ID', 'Product_ID', 'Shipping_ID', 'Quantity', 'Total_Price', 'Order_Date', 'Payment_Method']]
order_tb.head()
```

```python
# Yanki Table
yanki_tb = pd.merge(yanki_data[['Order_ID', 'Customer_ID', 'Product_ID', 'Transaction_Status']], shipping_tb[['Customer_ID', 'Shipping_ID']], on='Customer_ID').copy().drop_duplicates().reset_index(drop=True)
yanki_tb.head()
```

```python
# Save the files to csv
customer_tb.to_csv(r'Dataset\clean_data\customer_tb.csv', index=False)
product_tb.to_csv(r'Dataset\clean_data\product_tb.csv', index=False)
shipping_tb.to_csv(r'Dataset\clean_data\shipping_tb.csv', index=False)
order_tb.to_csv(r'Dataset\clean_data\order_tb.csv', index=False)
yanki_tb.to_csv(r'Dataset\clean_data\yanki_tb.csv', index=False)
```

```python
# Create schema and table

def create_tables():
    conn = get_db_connection()
    cur = conn.cursor()
    create_table_query = '''
        CREATE SCHEMA IF NOT EXISTS yanki_ecommerce_db;

        DROP TABLE IF EXISTS yanki_ecommerce_db.CUSTOMER CASCADE;
        DROP TABLE IF EXISTS yanki_ecommerce_db.PRODUCT CASCADE;
        DROP TABLE IF EXISTS yanki_ecommerce_db.SHIPPING CASCADE;
        DROP TABLE IF EXISTS yanki_ecommerce_db.ORDER CASCADE;
        DROP TABLE IF EXISTS yanki_ecommerce_db.YANKI_TB CASCADE;

        CREATE TABLE IF NOT EXISTS yanki_ecommerce_db.CUSTOMER (
            CUSTOMER_ID UUID PRIMARY KEY,
            FIRST_NAME VARCHAR(25) NOT NULL,
            LAST_NAME VARCHAR(25) NOT NULL,
            EMAIL VARCHAR(70),
            PHONE_NUM VARCHAR(40)
        );

        CREATE TABLE IF NOT EXISTS yanki_ecommerce_db.PRODUCT (
            PRODUCT_ID UUID PRIMARY KEY,
            PRD_NAME VARCHAR(50) NOT NULL,
            BRAND VARCHAR(50) NOT NULL,
            CATEGORY VARCHAR(70) NOT NULL,
            PRICE DECIMAL(5,2) NOT NULL
        );

        CREATE TABLE IF NOT EXISTS yanki_ecommerce_db.SHIPPING (
            SHIPPN_ID SERIAL PRIMARY KEY,
            CUSTOMER_ID UUID NOT NULL,
            POST_CODE VARCHAR(100) NOT NULL,
            CITY VARCHAR(100) NOT NULL,
            STATE VARCHAR(100) NOT NULL,
            COUNTRY VARCHAR(100) NOT NULL,
            SHIPPN_ADDR VARCHAR(150) NOT NULL,

            FOREIGN KEY (CUSTOMER_ID) REFERENCES yanki_ecommerce_db.CUSTOMER (CUSTOMER_ID)
        );

        CREATE TABLE IF NOT EXISTS yanki_ecommerce_db.ORDER (
            ORDER_ID UUID PRIMARY KEY,
            CUSTOMER_ID UUID NOT NULL,
            PRODUCT_ID UUID NOT NULL,
            SHIPPN_ID INT NOT NULL,
            QUANTITY INT NOT NULL,
            TOTAL_PRC DECIMAL(10, 2) NOT NULL,
            ORDER_DATE TIMESTAMP NOT NULL,
            PYMT_MTHD VARCHAR(30) NOT NULL,

            FOREIGN KEY (CUSTOMER_ID) REFERENCES yanki_ecommerce_db.CUSTOMER (CUSTOMER_ID),
            FOREIGN KEY (PRODUCT_ID) REFERENCES yanki_ecommerce_db.PRODUCT (PRODUCT_ID),
            FOREIGN KEY (SHIPPN_ID) REFERENCES yanki_ecommerce_db.SHIPPING (SHIPPN_ID)
        );

        CREATE TABLE IF NOT EXISTS yanki_ecommerce_db.YANKI_TB (
            ORDER_ID UUID NOT NULL,
            CUSTOMER_ID UUID NOT NULL,
            PRODUCT_ID UUID NOT NULL,
            TRNSTN_STATUS VARCHAR(30) NOT NULL,
            SHIPPN_ID INT NOT NULL,
            
            FOREIGN KEY (ORDER_ID) REFERENCES yanki_ecommerce_db.ORDER (ORDER_ID),
            FOREIGN KEY (CUSTOMER_ID) REFERENCES yanki_ecommerce_db.CUSTOMER (CUSTOMER_ID),
            FOREIGN KEY (PRODUCT_ID) REFERENCES yanki_ecommerce_db.PRODUCT (PRODUCT_ID),
            FOREIGN KEY (SHIPPN_ID) REFERENCES yanki_ecommerce_db.SHIPPING (SHIPPN_ID)
        );
        '''
    cur.execute(create_table_query)
    conn.commit()
    cur.close()
    conn.close()
```

```python
# Create the tables
create_tables()
```

```python
# loading data for CUSTOMER table
import csv

def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cur = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            cur.execute(
                '''
                INSERT INTO yanki_ecommerce_db.CUSTOMER(CUSTOMER_ID, FIRST_NAME, LAST_NAME, EMAIL, PHONE_NUM)
                VALUES (%s, %s, %s, %s, %s);
                ''',
                row
            )
    conn.commit()
    cur.close()
    conn.close()

# Define csv file path
csv_file_path = r'Dataset\clean_data\customer_tb.csv'

load_data_from_csv(csv_file_path)
print('CUSTOMER date loaded successfully')
```
